# Access to Jobs, Roads/Transit

```python
# Read in data to calculate ATO Scores by City Area

# read in TAZ only for WFRC/MAG
sdfTAZ = pd.DataFrame.spatial.from_featureclass(strTAZIn)
sdfTAZ = sdfTAZ[sdfTAZ['SUBAREAID']==1]

# read in TDM ATO output
dfAtoWfrc = pd.read_csv(strAtoWfrc)

# merge by CO_TAZID
sdfTAZwithATO = pd.DataFrame.merge(sdfTAZ,dfAtoWfrc,on='CO_TAZID',how='left')

display(sdfTAZwithATO)
display(sdfTAZwithATO.columns)
```

```python
display(sdfTAZwithATO.groupby(['CITY_NAME']).agg(CNTREC=('CO_TAZID','count')))
```

```python
sdfTAZwithATO.columns
```

```python
sdfTAZwithATO['TAZID'] = sdfTAZwithATO['TAZID_x']
sdfTAZwithATO[['TAZID','AUTO_JB','SHAPE']].spatial.to_featureclass(os.path.join(strGDB, r"AccessibleJobsAuto"),sanitize_columns=False)
sdfTAZwithATO[['TAZID','TRAN_JB','SHAPE']].spatial.to_featureclass(os.path.join(strGDB, r"AccessibleJobsTransit"),sanitize_columns=False)
```

```python
# use zeros where no jobs
sdfTAZwithATO['AUTO_JB'].fillna(0, inplace=True)
sdfTAZwithATO['TRAN_JB'].fillna(0, inplace=True)

dfATOCalcs_CityMax = sdfTAZwithATO.groupby(['CITY_NAME'],as_index=False).agg(CNTREC=('CO_TAZID','count'),AUTO_JB_max=('AUTO_JB','max'),TRAN_JB_max=('TRAN_JB','max'))
display(dfATOCalcs_CityMax)
```

```python
sdfATOCalcs_PercentMax = pd.DataFrame.merge(sdfTAZwithATO, dfATOCalcs_CityMax, on='CITY_NAME')
sdfATOCalcs_PercentMax['AUTO_JB_percentCityMax'] = sdfATOCalcs_PercentMax['AUTO_JB'] / sdfATOCalcs_PercentMax['AUTO_JB_max']
sdfATOCalcs_PercentMax['TRAN_JB_percentCityMax'] = sdfATOCalcs_PercentMax['TRAN_JB'] / sdfATOCalcs_PercentMax['TRAN_JB_max']
sdfATOCalcs_PercentMax = sdfATOCalcs_PercentMax.fillna(0)
sdfATOCalcs_PercentMax
```

```python
# check max scores for all cities to make sure equals 1
dfCheck = sdfATOCalcs_PercentMax.groupby(['CITY_NAME'],as_index=False).agg(NumTAZ=('CO_TAZID','count'),AutoMaxScore=('AUTO_JB_percentCityMax','max'),TranMaxScore=('TRAN_JB_percentCityMax','max'),AutoMinScore=('AUTO_JB_percentCityMax','min'),TranMinScore=('TRAN_JB_percentCityMax','min'))
dfCheck
```

```python
sdfATOScores = sdfATOCalcs_PercentMax[['CO_TAZID','AUTO_JB_percentCityMax','TRAN_JB_percentCityMax','SHAPE']]
sdfATOScores = sdfATOScores.rename(columns={'AUTO_JB_percentCityMax':'ATO_Auto_Score','TRAN_JB_percentCityMax':'ATO_Tran_Score'})
sdfATOScores.spatial.to_featureclass(strTAZwithATOOut,sanitize_columns=False)
```

# Ring Buffers

```python
def getLayerDistField(strName):
  return "dist_" + strName.rsplit('\\',1)[1]
```

```python
def createRingBuffers(strBufferPrefix,dBufferDistances):
  strBufferLayerName = strBufferPrefix  + "_Buffers"
  deleteIfExists(strBufferLayerName)
  deleteIfExists("memory\diss_fc")
  display(strBufferPrefix  + "_Buffers")
  arcpy.analysis.MultipleRingBuffer(strBufferPrefix, strBufferLayerName, dBufferDistances, "miles", "", "ALL")
  strNewFieldName = getLayerDistField(strBufferPrefix)
  arcpy.management.AlterField(strBufferLayerName, "distance", strNewFieldName, strNewFieldName)
```

```python
#indicate which layers to buffer by putting "Yes" in second column

# TEMP FOR GROCERY ONLY

dBufferLayers = [
  [strInterchangesOut_Cur , "No", [1/2, 1, 2]],
  [strInterchangesOut_Fut , "No", [1/2, 1, 2]],
  [strTransitOut_Lcl      , "No", [1/4      ]],
  [strTransitOut_BrtCur   , "No", [1/3      ]],
  [strTransitOut_BrtFut   , "No", [1/3      ]],
  [strTransitOut_LrtCur   , "No", [1/3      ]],
  [strTransitOut_LrtFut   , "No", [1/3      ]],
  [strTransitOut_CrtCur   , "No", [1/3      ]],
  [strTransitOut_CrtFut   , "No", [1/3      ]],
  [strChildCareOut        , "No", [1/2, 1   ]],
  [strHealthCareOut       , "No", [1/2, 1   ]],
  [strSchoolsOut_RegPub   , "No", [1/2, 1   ]],
  [strSchoolsOut_HighEd   , "No", [     1, 2]],
  [strGroceryOut          , "Yes", [1/2, 1   ]],
  [strCommunityCenterOut  , "No", [1/2, 1   ]],
  [strPathsOut_Cur        , "No", [1/2, 1   ]],
  [strPathsOut_Fut        , "No", [1/2, 1   ]],
  [strATCycleTracksOut_Cur, "No", [1/2, 1   ]],
  [strATCycleTracksOut_Fut, "No", [1/2, 1   ]]
]

for l in dBufferLayers:
  if (l[1])=='Yes':
    print('buffering ' + l[0])
    createRingBuffers(l[0],l[2])
print('done!')
```
